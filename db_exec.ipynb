```python
import sqlite3
import pandas as pd

# Path to your DB
db_path = "/Users/swehavenkateshwari/Documents/Github/Alzheimer_AI_Diagnosis/Alzheimer_Project/Alzheimer_Database/alzheimer_predictions.db"

# Connect to the database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Read global_importance table
global_df = pd.read_sql_query("SELECT * FROM global_importance", conn)

# Read individual_predictions table
individual_df = pd.read_sql_query("SELECT * FROM individual_predictions", conn)

batch_pred = pd.read_sql_query("SELECT * FROM batch_predictions", conn)

region_df = pd.read_sql_query("SELECT * FROM batch_region", conn)

image_df = pd.read_sql_query("SELECT * FROM stored_images", conn)


# Display
print("=== Global Importance ===")
print(global_df.head())

print("\n=== Individual Predictions ===")
print(individual_df.head())

print("\n=== Batch Predictions ===")
print(batch_pred.head())
print("\n=== Region Predictions ===")
print(region_df.head())
print("\n=== Stored Images ===")
print(image_df.head(50))



# Count rows in global_importance
cursor.execute("SELECT COUNT(*) FROM global_importance")
global_count = cursor.fetchone()[0]

# Count rows in individual_predictions
cursor.execute("SELECT COUNT(*) FROM individual_predictions")
individual_count = cursor.fetchone()[0]

# Count rows in batch_predictions
cursor.execute("SELECT COUNT(*) FROM batch_predictions")
batch_count = cursor.fetchone()[0]
# Count rows in region_predictions
cursor.execute("SELECT COUNT(*) FROM batch_region")
region_count = cursor.fetchone()[0]
# Print counts
print(f"Rows in global_importance: {global_count}")
print(f"Rows in individual_predictions: {individual_count}")
print(f"Rows in batch_predictions: {batch_count}")
print(f"Rows in region_predictions: {region_count}")
# Count rows in batch_predictions
print("\n=== Batch Predictions ===")

# Close
conn.close()


```

```python
from alzheimers_db_setup import AlzheimerPredictionStorage

# Check what's in the database
storage = AlzheimerPredictionStorage()

print("=== Database Contents ===")

# Check batch_predictions
print("\n1. Batch Predictions:")
batch_preds = storage.get_batch_predictions()
print(f"   Total rows: {len(batch_preds)}")
if batch_preds:
    print("   First row:")
    for key, value in batch_preds[0].items():
        print(f"      {key}: {value}")

# Check batch_region
print("\n2. Batch Regions:")
regions = storage.get_batch_region()
print(f"   Total rows: {len(regions)}")
if regions:
    print("   Unique filenames:", set([r['Filename'] for r in regions]))
    print("   Brain regions:", set([r['Brain_Region'] for r in regions]))

# Check stored_images
print("\n3. Stored Images:")
images = storage.get_stored_images()
print(f"   Total rows: {len(images)}")
if images:
    print("   Patient IDs:", set([img['patient_id'] for img in images]))
    print("   Image types:", set([img['image_type'] for img in images]))

# Get statistics
print("\n4. Statistics:")
stats = storage.get_statistics()
print(f"   Batch predictions: {stats['batch_predictions']['total_batch_predictions']}")
print(f"   Class distribution: {stats['batch_predictions']['class_distribution']}")

storage.close()
```
