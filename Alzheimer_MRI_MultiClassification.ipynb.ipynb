# Alzheimer Diease Prediction using Deep Learning
## MRI Image Pipeline for Alzheimer's Disease Diagnosis
## Implements preprocessing, InceptionV3-based classification, and Grad-CAM interpretability

```python
# 📦 Import necessary libraries
import os  # For handling directories
import numpy as np  # For numerical operations
import tensorflow as tf  # Core deep learning framework
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # For image augmentation
from tensorflow.keras import layers, models  # For building neural networks
from tensorflow.keras.applications import InceptionV3  # Pre-trained CNN model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau  # For training control
import matplotlib.pyplot as plt  # For plotting
from sklearn.metrics import classification_report, confusion_matrix, f1_score, cohen_kappa_score  # Evaluation metrics
import time  # For measuring training duration
import json  # For saving results if needed

# 🗂 Set directory paths for dataset and output
DATA_DIR = '/Users/swehavenkateshwari/F416664_Alzheimers_AI_Diagnosis/MRI_Image_Data'
OUTPUT_DIR = '/Users/swehavenkateshwari/F416664_Alzheimers_AI_Diagnosis/MRI_Image_Data/Output_Results'

# 🖼 Image and training parameters
IMG_SIZE = 331  # InceptionV3 expects minimum 299x299, slightly larger used here
BATCH_SIZE = 16
EPOCHS = 50
LEARNING_RATE = 0.001

# Define class names for Alzheimer's stages
CLASS_NAMES = ['Mild_Demented', 'Moderate_Demented', 'Non_Demented', 'Very_Mild_Demented']

# ✅ Create output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 🔄 Data augmentation for training images
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    rotation_range=20,  # Random rotation
    width_shift_range=0.1,  # Horizontal shift
    height_shift_range=0.1,  # Vertical shift
    shear_range=0.1,  # Shear transformation
    zoom_range=0.1,  # Random zoom
    horizontal_flip=True,  # Horizontal flip
    fill_mode='nearest',  # Fill missing pixels
    validation_split=0.2  # Reserve 20% for validation
)

# ✅ Validation generator (no augmentation, only rescaling)
val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Image size tuple for generator
image_size = (IMG_SIZE, IMG_SIZE)

# 🚀 Create training generator
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=image_size,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42
)

# 🚀 Create validation generator
validation_generator = val_datagen.flow_from_directory(
    DATA_DIR,
    target_size=image_size,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=42
)

# ℹ Print dataset statistics
print(f"Training samples: {train_generator.samples}")
print(f"Validation samples: {validation_generator.samples}")
print(f"Class indices: {train_generator.class_indices}")

# 🏗 Build model using Functional API for better Grad-CAM access
def build_functional_model(num_classes=4):
    """Build Alzheimer’s classifier using pre-trained InceptionV3 and custom layers"""
    
    # Input layer
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name='input_layer')
    
    # Load pre-trained InceptionV3 without top layers
    base_model = InceptionV3(
        weights='imagenet',
        include_top=False,
        input_tensor=inputs,
        pooling=None
    )
    
    # Freeze base model initially
    base_model.trainable = False
    
    # Get output from base model
    x = base_model(inputs, training=False)
    
    # Add global pooling and batch normalization
    x = layers.GlobalAveragePooling2D(name='global_avg_pool')(x)
    x = layers.BatchNormalization(name='batch_norm_1')(x)
    
    # Add dense layers with dropout for regularization
    x = layers.Dense(1024, activation='relu', name='dense_1')(x)
    x = layers.Dropout(0.5, name='dropout_1')(x)
    
    x = layers.Dense(512, activation='relu', name='dense_2')(x)
    x = layers.Dropout(0.5, name='dropout_2')(x)
    
    x = layers.Dense(256, activation='relu', name='dense_3')(x)
    x = layers.Dropout(0.4, name='dropout_3')(x)
    
    # Output layer with softmax
    outputs = layers.Dense(num_classes, activation='softmax', name='output_layer')(x)
    
    # Create functional model
    model = models.Model(inputs=inputs, outputs=outputs, name='alzheimer_classifier')
    
    # Unfreeze last few layers of base model for fine-tuning
    for layer in base_model.layers[-20:]:
        layer.trainable = True
    
    return model, base_model

# 🔨 Build the model
print("\nBuilding model...")
model, base_model = build_functional_model(num_classes=len(CLASS_NAMES))

# ⚡ Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# 📝 Print model summary
print("\nModel Summary:")
model.summary()

# ⏱ Callbacks for early stopping and learning rate reduction
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=7,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]

# Record start time
start_time = time.time()

# 🏋️ Train the model
print("\nStarting training...")
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    callbacks=callbacks,
    verbose=1
)

# Calculate training duration
training_time = time.time() - start_time
hours, remainder = divmod(training_time, 3600)
minutes, seconds = divmod(remainder, 60)

print(f"\nTraining completed!")
print(f"Training Time: {int(hours)} hours, {int(minutes)} minutes, {int(seconds)} seconds")

# 📊 Evaluate model performance
print("\nEvaluating model...")
train_metrics = model.evaluate(train_generator, verbose=0)
val_metrics = model.evaluate(validation_generator, verbose=0)

print(f"\nResults:")
print(f"Training - Loss: {train_metrics[0]:.4f}, Accuracy: {train_metrics[1]*100:.2f}%, AUC: {train_metrics[2]:.4f}")
print(f"Validation - Loss: {val_metrics[0]:.4f}, Accuracy: {val_metrics[1]*100:.2f}%, AUC: {val_metrics[2]:.4f}")

# 🔮 Generate predictions on validation set
print("\nGenerating predictions...")
validation_generator.reset()
predictions = model.predict(validation_generator, verbose=1)
y_pred = np.argmax(predictions, axis=1)
y_true = validation_generator.classes[:len(y_pred)]

# Ensure class names are in correct order
class_names_ordered = [k for k, v in sorted(validation_generator.class_indices.items(), key=lambda item: item[1])]

# 📈 Calculate metrics
cm = confusion_matrix(y_true, y_pred)
report = classification_report(y_true, y_pred, target_names=class_names_ordered, output_dict=True)
f1_weighted = f1_score(y_true, y_pred, average='weighted')
kappa = cohen_kappa_score(y_true, y_pred)

print('\nClassification Report:')
print(classification_report(y_true, y_pred, target_names=class_names_ordered))
print(f"\nWeighted F1 Score: {f1_weighted:.4f}")
print(f"Cohen's Kappa Score: {kappa:.4f}")

# 📊 Plot training and evaluation results
fig = plt.figure(figsize=(20, 12))

# 1️⃣ Accuracy curve
plt.subplot(2, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('Model Accuracy', fontsize=14)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# 2️⃣ Loss curve
plt.subplot(2, 3, 2)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Model Loss', fontsize=14)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# 3️⃣ Confusion matrix
plt.subplot(2, 3, 3)
im = plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix', fontsize=14)
plt.colorbar(im, fraction=0.046, pad=0.04)
tick_marks = np.arange(len(class_names_ordered))
plt.xticks(tick_marks, class_names_ordered, rotation=45, ha='right')
plt.yticks(tick_marks, class_names_ordered)
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# 4️⃣ AUC curve
plt.subplot(2, 3, 4)
plt.plot(history.history['auc'], label='Training AUC', linewidth=2)
plt.plot(history.history['val_auc'], label='Validation AUC', linewidth=2)
plt.title('Model AUC', fontsize=14)
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.legend()
plt.grid(True, alpha=0.3)

# 5️⃣ Per-class metrics
plt.subplot(2, 3, 5)
class_f1_scores = [report[class_name]['f1-score'] for class_name in class_names_ordered]
class_precisions = [report[class_name]['precision'] for class_name in class_names_ordered]
class_recalls = [report[class_name]['recall'] for class_name in class_names_ordered]

x = np.arange(len(class_names_ordered))
width = 0.25

plt.bar(x - width, class_precisions, width, label='Precision', alpha=0.8)
plt.bar(x, class_recalls, width, label='Recall', alpha=0.8)
plt.bar(x + width, class_f1_scores, width, label='F1-Score', alpha=0.8)

plt.xlabel('Classes')
plt.ylabel('Score')
plt.title('Per-Class Performance Metrics', fontsize=14)
plt.xticks(x, class_names_ordered, rotation=45, ha='right')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

# 6️⃣ Learning rate plot or summary if not available
plt.subplot(2, 3, 6)
if 'lr' in history.history:
    plt.plot(history.history['lr'], linewidth=2)
    plt.title('Learning Rate Schedule', fontsize=14)
    plt.xlabel('Epochs')
    plt.ylabel('Learning Rate')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
else:
    plt.text(0.1, 0.9, 'Model Summary', fontsize=16, weight='bold', transform=plt.gca().transAxes)
    summary_text = f"""
Total Parameters: {model.count_params():,}
Trainable Parameters: {sum(1 for p in model.trainable_weights):,}
Non-trainable Parameters: {model.count_params() - sum(1 for p in model.trainable_weights):,}

Best Validation Accuracy: {max(history.history['val_accuracy'])*100:.2f}%
Final Validation Accuracy: {val_metrics[1]*100:.2f}%
Weighted F1 Score: {f1_weighted:.4f}
Cohen's Kappa: {kappa:.4f}

Training Time: {int(hours)}h {int(minutes)}m {int(seconds)}s
    """
    plt.text(0.1, 0.7, summary_text, fontsize=12, transform=plt.gca().transAxes, 
             verticalalignment='top', family='monospace')
    plt.axis('off')

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'training_results.png'), dpi=300, bbox_inches='tight')
plt.show()

# 🏁 Print training summary
print("\n" + "="*60)
print("TRAINING SUMMARY")
print("="*60)
print(f"Model Architecture: InceptionV3 + Custom Layers")
print(f"Total Classes: {len(CLASS_NAMES)}")
print(f"Training Time: {int(hours)}h {int(minutes)}m {int(seconds)}s")
print(f"Best Validation Accuracy: {max(history.history['val_accuracy']) * 100:.2f}%")
print(f"Final Validation Accuracy: {val_metrics[1] * 100:.2f}%")
print(f"Weighted F1 Score: {f1_weighted:.4f}")
print(f"Cohen's Kappa Score: {kappa:.4f}")
print("="*60)
```
